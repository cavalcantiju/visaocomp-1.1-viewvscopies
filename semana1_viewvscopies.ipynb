```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Criar imagem preta 100x100
imagem = np.zeros((100, 100), dtype=np.uint8)

# 2. Criar a variável "recorte" (quadrado central 20x20)
# O centro de 100 é 50. Para um quadrado de 20, pegamos de 40 a 60.
recorte = imagem[40:60, 40:60]

# 3. Pintar o recorte de branco
recorte[:] = 255

# 4. Mostrar a imagem original
plt.imshow(imagem, cmap='gray')
plt.title("Imagem Original (com alteração via View)")
plt.show()
```

# O que aconteceu no código:

Economia de Memória: Quando crio a var recorte, eu não dupliquei os pixels. Eu apenas criei um "ponteiro inteligente". Basicamente, eu disse para o Python: "Olha para a imagem original, mas foca só nesse pedaço aqui".

Parece uma cópia, mas como é uma View, se eu pinto o recorte de branco, eu estou mexendo direto nos endereços de memória da imagem original.

O Resultado: Por isso que, no final, quando mandamos mostrar a imagem original, ela aparece com o quadrado branco no meio. Eu nunca dei um comando direto na imagem para pintar o centro, eu fiz isso através da variável recorte.
